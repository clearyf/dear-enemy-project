```python
def do_boxplot(data, name):
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    props = {'connectionstyle':patches.ConnectionStyle('bar', fraction=0.3),\
             'arrowstyle':'-','shrinkA':10,'shrinkB':10,'linewidth':1}
    fig, ax = plt.subplots()
    height = max(max(data['near_s']), max(data['far_s']))
    if name == 'habituation_treatment' or name == 'experiment_control':
        ax.annotate('', xy=(5,height), xytext=(6,height), arrowprops=props)
        ax.annotate('*', xy=(5.5,height+13))

    ax.boxplot(
        data.values(),
        labels=data.keys()
        # positions=[1, 3, 3.75, 5.25, 6, 8]
    )
    ax.set(ylabel='attacks per 2 min')
    ax.set_ylim(0, 150)
    #print(f'{ax.get_ybound()}')
    #(lower, upper) = ax.get_ybound()
    #ax.set_ybound((lower, upper + 20))

    fig.savefig(f'intruder_position_boxplots_{name}.pdf', format='pdf')
```
